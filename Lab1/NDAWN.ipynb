```python
#The provided code sends an HTTP GET request to the 'dataurl' URL, retrieves CSV data from the NDAWN website for specified weather stations and date range, and checks for any HTTP response errors while disabling SSL certificate verification
dataurl = r'https://ndawn.ndsu.nodak.edu/table.csv?ttype=sbgdd&station=162&station=98&year=2023&begin_date=2023-05-01&end_date=2023-09-15'
NDAWN_Geo_Get = requests.get(dataurl, verify=False, stream=True) #get request funciton with url
NDAWN_Geo_Get.raise_for_status()

print(NDAWN_Geo_Get)
#creates folder
import os
##print(os.getcwd())
NDAWN_Geofolder_new = r'C:\Users\Track\OneDrive\Documents\ArcGIS\Projects\Lab_1\NDAWN_DOGG'
if not os.path.exists(NDAWN_Geofolder_new):
    os.mkdir(NDAWN_Geofolder_new)
```

```python
#saving the downloaded data to the local file system.
with open(NDAWN_Geofolder_new+'\\test.csv', 'wb') as file:
    file.write(NDAWN_Geo_Get.content)
```

```python
#Clean Data
NDAWNoutput = r'C:\Users\Track\OneDrive\Documents\ArcGIS\Projects\Lab_1\NDAWN_DOGG.csv'
NDAWNresponse = requests.get(dataurl, stream=True)

#open empty file to write
with open (NDAWNoutput, 'wb') as file:
    file.write(NDAWNresponse.content)
#The provided code reads a CSV file ('NDAWN_DOGG.csv') into a pandas DataFrame, skips the first 4 rows in the CSV file, assigns new column names to the DataFrame, and displays the DataFrame with the updated column names.
North_Dogg_Dataframe = pd.read_csv(r'C:\Users\Track\OneDrive\Documents\ArcGIS\Projects\Lab_1\NDAWN_DOGG.csv', skiprows=4)
new_names = ['Station Name', 'Latitude', 'Longitude', 'Elevation', 'Year', 'Month', 'Day', 'Max Temp', 'Max Temp Flag', 'Min Temp', 'Min Temp Flag', 'Rainfall', 'Rainfall Flag', 'Sugarbeet Daily Growing Degree Days', 'Sugarbeet Daily Growing Degree Days Flag', 'Sugarbeet Accumulated Growing Degree Days', 'Sugarbeet Accumulated Growing Degree Days Flag', 'Sugarbeet Estimated Growth Stage']
North_Dogg_Dataframe.columns = new_names
North_Dogg_Dataframe
```

```python
# Define Coordinate System WGS84)
ndgeometry = [Point(x,y) for x,y in zip(North_Dogg_Dataframe['Longitude'], North_Dogg_Dataframe['Latitude'])]
ndgeoframe = gpd.GeoDataFrame(North_Dogg_Dataframe, geometry=ndgeometry, crs = 'WGS84')
#print(ndgeoframe)
```

```python
# Define the output Shapefile name
output_shapefile = 'NDAWNbb.shp'  # Replace with the desired output file name
# Save the GeoDataFrame as a Shapefile
ndgeoframe.to_file(NDAWN_Geofolder_new+ r'\NDAWNbb.shp', driver='ESRI Shapefile')

```
